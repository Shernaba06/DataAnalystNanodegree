```python
y = y.cumsum()
plt.plot(y)
```

We cain either plot two arrays in the same plot like in the example above or we can define subplots and plot each array in a separate subplot

```python
plt.subplot(1,2,1)
plt.plot(x,y, marker='o', color='red', alpha=0.5)

plt.subplot(1,2,2)
plt.plot(x,y2, marker='o', color='black', alpha=0.5)
;
```

Subplots does not necessarily need to be side by side, a plot can be nested inside other plot. By using the fig.add_axes parameters we can define the size of the big, main plot (ax1 in the example below), and the size of the small, nested plot (ax2 in the example below). Notice that the size of the 

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.6,0.4,0.3])

ax1.plot(x,y,'r')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Y plot')

ax2.plot(x,y2,'g')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Y2 plot')
;
```
